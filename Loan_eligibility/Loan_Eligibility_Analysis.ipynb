## Data Transformations

Dropping CNT_CHILDREN and AMT_GOODS_PRICE has they have high correlation with CNT_FAM_MEMBERS and AMT_CREDIT columns respectively.

We had seggregated the features into numericalCols and categoricalColumns for each of them to go through few transformations prior getting added to Vector Assembler transformation.

```
# Dropping the rows with empty entries.
# Categorizing columns into Categorical and Numerical Columns.

revisedDataFrame=revisedDataFrame.drop('CNT_CHILDREN')
finalDataFrame=revisedDataFrame.drop('AMT_GOODS_PRICE')


numericalCols = [item[0] for item in finalDataFrame.dtypes if item[1].startswith('int') | item[1].startswith('double')][1:]

categoricalCols = [item[0] for item in finalDataFrame.dtypes if item[1].startswith('string')] 
print(str(len(numericalCols)) + '  numerical features')
print(str(len(categoricalCols)) + '  categorical features')
print("Total rows in the dataset is: ", finalDataFrame.count())
```

Categorical Columns: 
To convert them to numerical Columns by adding dummy fields, they go through transformations like StringIndexer, Onehot encoder estimator.

Numerical Columns: 
They go through impute transformation to fill all the empty entries with mean/median of non-empty enties.

The final transformed stages are then added to vector assembler stage, which results in vectorized version of all the features.

We use pipelines as they are Syntactic sugar for transformations, We add these transformation into pipelines.

```
# Converting the Categorial columns to numerical columns and adding dummy fields.
# Pipelining the different transformation steps into a single one to be run on Categorical columns.
# Sending all transformed columns through vectorassembler to create a feature called "features".

from pyspark.ml.feature import  VectorAssembler,OneHotEncoderEstimator, StringIndexer, Imputer
from pyspark.ml import Pipeline, PipelineModel

pipeline=Pipeline(stages=[]);
commonPipeline=[];
indexerOneHotPipeline=[]
imputePipeline=[]
for categoricalCol in categoricalCols:
    stringIndexer = StringIndexer(inputCol = categoricalCol, outputCol = categoricalCol + "Index")
    encoder = OneHotEncoderEstimator(inputCols=[categoricalCol + "Index"], outputCols=[categoricalCol + "Encoded"])
    commonPipeline += [stringIndexer, encoder]
    indexerOneHotPipeline += [stringIndexer, encoder]
    
    
for numerical in numericalCols:
    imputer=Imputer(inputCols=[numerical], outputCols=[numerical+"Imputed"])
    imputePipeline+=[imputer]
  
  
assemblerInputs = [c +"Encoded" for c in categoricalCols]
assemblerInputs+= [c+"Imputed" for c in numericalCols]
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
commonPipeline += [assembler]
```

We process the numerical Columns through impute transformatione, as we would be dropping rows with empty entries.

```
imputePipelineFinal=Pipeline(stages=imputePipeline)
finalDataFrame=imputePipelineFinal.fit(finalDataFrame).transform(finalDataFrame);
```

```
# Drop the rows with empty entries in any of the columns.
finalDataFrame=finalDataFrame.dropna()
```

We are fitting and tranforming the data through this transformation prior to cross validation stage is because, because in cross validation if we have any features in validation set, which are not in training set, then we face errors. So, we transform data here itself.

```
# Since, we think there are issues placing Indexer, OneHotEncoder and Vectoraseembler in the same pipeline. We break them into two steps.
# In the first step, we are working on Indexer, OneHotEncoder and transforming the data accordingly.


indexerOneHotPipe=Pipeline(stages=indexerOneHotPipeline)
finalDataFrame=indexerOneHotPipe.fit(finalDataFrame).transform(finalDataFrame)
display(finalDataFrame.head(10))
```

## Logistic Regression

```
# Adding the Models to be compared to the Assembler pipeline.
# We are using Logistic Regression.

from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier, NaiveBayes
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import BinaryClassificationEvaluator


#Logistic Regression.
lr=LogisticRegression(maxIter=10)
pipeLineLogistic=[assembler,lr]
paramGridLogistic = ParamGridBuilder()\
          .baseOn({pipeline.stages: pipeLineLogistic})\
          .addGrid(lr.regParam,[0.01, .04])\
          .addGrid(lr.elasticNetParam,[0.1, 0.4])\
          .build()


# Appending all ParamGrid pipeline steps to push into Cross Validator.
totalParamGrid=paramGridLogistic;
```

## Random Forest Classifier

```
randomForest = RandomForestClassifier()
pipelineRandomForest = [assembler, randomForest]
paramGridRandomForest = ParamGridBuilder()\
      .baseOn({pipeline.stages: pipelineRandomForest})\
      .addGrid(randomForest.numTrees,[30,35])\
      .build()

 
# Appending all ParamGrid pipeline steps to push into Cross Validator.
totalParamGrid+=paramGridRandomForest
```

## Gradient Boost Classifier

```
# gradientBoost = GBTClassifier(labelCol="label", featuresCol="features",  maxBins=42,  maxDepth=10, maxIter=15)
gradientBoost = GBTClassifier(labelCol="label", featuresCol="features", maxIter=15)
pipelineGradientBoost = [assembler, gradientBoost]
paramGridGradientBoost = ParamGridBuilder()\
      .baseOn({pipeline.stages: pipelineGradientBoost})\
      .addGrid(gradientBoost.maxBins,[40,44])\
      .addGrid(gradientBoost.maxDepth,[6,10])\
      .build()


# Appending all ParamGrid pipeline steps to push into Cross Validator.
totalParamGrid+=paramGridGradientBoost
```

## Splitting data into Train and Test Data.

```
# Splitting the Transformed data to Train and Test(Validation) data.
trainingData, testingData=finalDataFrame.randomSplit([0.7,0.3],43)
print("Training Data rows: ", trainingData.count())
print("Validation(Test) Data rows: ", testingData.count())
```

## Cross Validator

```
#CrossValidator, uses Binary Class Evaluator since we are performing Binary Classification.
# We fit the model to the transformed Data and Cross validate it on given number of folds.

from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import BinaryClassificationEvaluator

cv = CrossValidator()\
      .setEstimator(pipeline)\
      .setEvaluator(BinaryClassificationEvaluator())\
      .setEstimatorParamMaps(totalParamGrid)\
      .setNumFolds(5)

cvModel = cv.fit(trainingData)
```

```
# Getting the metrics for the best model.
import numpy as np
# BinaryClassificationEvaluator defaults to ROC AUC, so higher is better
cvModel.getEstimatorParamMaps()[ np.argmax(cvModel.avgMetrics) ]
```

```
#Getting the metrics for the Worst Model.

cvModel.getEstimatorParamMaps()[ np.argmin(cvModel.avgMetrics) ]
```

```
# Function gives the metrics corresponding to models.

import re
def paramGrid_model_name(model):
  params = [v for v in model.values() if type(v) is not list]
  name = [v[-1] for v in model.values() if type(v) is list][0]
  name = re.match(r'([a-zA-Z]*)', str(name)).groups()[0]
  return "{}{}".format(name,params)

# Resulting metric and model description
# get the measure from the CrossValidator, cvModel.avgMetrics
# get the model name & params from the paramGrid
# put them together here:
kmeans_measures = zip(cvModel.avgMetrics, [paramGrid_model_name(m) for m in totalParamGrid])
metrics,model_names = zip(*kmeans_measures)
```

## Data Lineage:
We process the data through several steps of transformations:
Our intital data has 120 fields we picked 30 features which we feel are very much related to the final Loan eligibility Prediction.

We converted few integer columns to double columns as they would be going through impute transformation to fill the empty entries with non-empty entries.

We drew the heatmap with the shortlisted features, we drop those features(CNT_COUNT, AMT_GOODS_PRICE) which have maximum correlation with other features.

We seggregated the features into numerical and categorical columns, as each of them would be separate set of transformations prior getting added to vector assembler transformation.

Numerical columns go through impute transformation to fill the empty etnries with the mean/median of the non-empty entries. Once data goes through this transformation, we drop rows with empty entries.

Categorical Columns go through StringIndexer, OnehotEncoderEstimator transformations and gets converted to numerical columns by adding dummy columns.

We add these transformation stages into pipeline, as they are Syntactic sugar for processing transformations.

We split the data transformed through above transformations into training and test data.

We add the Vector assembler and Machine Learning model to the ParamGridBuilder for each of the models. We append the Param Grid Builder of each of the models onto single pipeline and pass it to Cross Validator as input pipeline. Once fit is triggered, it trains and validates all models on the range of parameters that we requested to test each of the models on, with the provided number of folds.

Cross validator returns the best and worst model parameters among all the provided models and the possible combinations of parameters that we requested to test the model on.
